**Pandas** :- Pandas is a fast, powerful, flexible and easy to use open source in Data analysis, and manipulation tool. Build on top of Python Programming language. This is know as a Pandas


**Syntax**:- *import pandas as pd*.


**"as pd"** is a alias/nick name/Short name.

```python
import pandas as pd
```

**Read the file**:-After creation of pandas and read the file inside the pandas.

**Syntax**:-  *read_**


**Star Symbol** can be indicate's type of the file like **CSV file**, **HTML file**, **SQL file**, etc.

```python
Pandasexample1 = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Titanic.csv')
```

**Pandas functions**:- (1) .head(), (2) .tail(), (3) .describe()

**.head()**:- it can be print the first five lines of the Data.

**.tail()**:- it will give the information of the data.

**.describe()**:- it will be dscribe the data.

```python
Pandasexample1.head()
```

```python
Pandasexample1.tail()
```

```python
Pandasexample1.describe()
```

```python
# display summary statistics of all columns
Pandasexample1.describe(include='all')
```

```python
# display column labels
Pandasexample1.columns
```

```python
# select a column as a series
Pandasexample1['sex']
```

```python
# unique categories in a column
Pandasexample1['sex'].unique()
```

```python
# number of unique categories in a column
Pandasexample1['sex'].nunique()
```

```python
# categories wise number of observations in a column
Pandasexample1['sex'].value_counts()
```

```python
# count of missing values
Pandasexample1.isna().sum()
```
